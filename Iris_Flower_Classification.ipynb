```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
iris = next(iter(uploaded))
iris_df = pd.read_csv(io.BytesIO(uploaded[iris]))
x = iris_df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalLengthCm']]
y = iris_df['Species']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
rf_classifier = RandomForestClassifier(n_estimators = 100,random_state = 42)
rf_classifier.fit(x_train,y_train)
y_pred = rf_classifier.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy:", accuracy)
conf_matrix = confusion_matrix(y_test,y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Reds", xticklabels=iris_df['Species'].unique(), yticklabels=iris_df['Species'].unique())
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show
```
